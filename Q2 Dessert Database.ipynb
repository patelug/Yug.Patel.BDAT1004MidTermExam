```python
#Question 2 Dessert Database

import sqlite3

conn = sqlite3.connect('dessert.db')
cursor = conn.cursor()

cursor.execute('''
    CREATE TABLE IF NOT EXISTS desserts (
        DessertName TEXT,
        ThomasRanking INTEGER,
        MyRanking INTEGER
    )
''')

desserts_data = [
    ('Brownie', 2, 3),
    ('Cake', 1, 5),
    ('Ice Cream', 3, 2),
    ('Pie', 5, 1),
    ('Cookies', 4, 4)
]

cursor.executemany('''
    INSERT INTO desserts (DessertName, ThomasRanking, MyRanking)
    VALUES (?, ?, ?)
''', desserts_data)

conn.commit()

# Query 1
cursor.execute('''
    SELECT * FROM desserts
    WHERE ThomasRanking = 1 OR MyRanking = 1
''')
print("Query 1:")
for row in cursor.fetchall():
    print(row)

# Query 2
cursor.execute('''
    SELECT * FROM desserts
    ORDER BY MyRanking ASC
''')
print("Query 2:")
for row in cursor.fetchall():
    print(row)

conn.close()
```
